```python
# for bar graph (visual) 
def changeTIPIValues(value):
    if value == 1:
        return 'Disagree strongly'
    if value == 2:
        return 'Disagree moderately'
    if value == 3:
        return 'Disagree a little'
    if value == 4:
        return 'Neither agree nor disagree'
    if value == 5:
        return 'Agree a little'
    if value == 6:
        return 'Agree moderately'
    if value == 7:
        return 'Agree strongly'

    return value

def changeFromToinTIPI(value, From, to):
    if value == From:
        return to
    return value
```

```python

for i in range(1,11):
  values = df[f'TIPI{i}'].value_counts().index.tolist()
  # maximum between {1,2,3,4,5,6 ... i.e chosen values in the dataset}
  maxima_value = max(values)

  # change 0 value to max value which is most chosen value in the survey
  df[f'TIPI{i}'] = df[f'TIPI{i}'].apply(lambda value: changeFromToinTIPI(value, 0, maxima_value))

  # convert numbers to string for better viz
  tipi = df[f'TIPI{i}'].apply(changeTIPIValues)

  plt.figure(figsize=(18, 7))
  sns.countplot(x=df[f'TIPI{i}'], hue=tipi)
```

## Categorizing Score

```python
df2['score'].describe()
```

Here,Mean is 100 so we take Moderate under 100


```python
df2[df2['score'] < 100]['score'].describe()
```

Likewise,We choose Mild under 77

```python
df2[df2['score'] < 77]['score'].describe()
```

We choose normal under 63

```python
df2[df2['score'] > 100]['score'].describe()
```

We choose severe as 123 and extremely severe above it

```python
def DepressionScaleGrouping(score):
    if score <= 63:
        return 'Normal'
    if 63 < score <= 77:
        return 'Mild'
    if 77 < score <= 100:
        return 'Moderate'
    if 100 < score <= 123:
        return 'Severe'
    if score > 123:
        return 'Extremely Severe'
df2['label'] = df2['score'].apply(DepressionScaleGrouping)
```

```python
sns.countplot(x=df2['label']);
```

## Balancing Data

Due to data imbalance lets move data 10 step forward

```python
def DepressionScaleGrouping(score):
    if score <= 73:
        return 'Normal'
    if 73 < score <= 87:
        return 'Mild'
    if 87 < score <= 110:
        return 'Moderate'
    if 110 < score <= 133:
        return 'Severe'
    if score > 133:
        return 'Extremely Severe'
df2['label'] = df2['score'].apply(DepressionScaleGrouping)
```

```python
sns.countplot(x=df2['label']);
```

## Splitting Data Set And Scaling

```python
label = df2['label']
df2.drop(['label',], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df2, label, test_size=.2)

print(f'x_train: {x_train.shape}, y_train: {y_train.shape}')
print(f'x_test: {x_test.shape}, y_test: {y_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Models

### Support Vector Machines (SVM)

```python
%%time
from sklearn.svm import SVC
from sklearn.metrics import classification_report

svc_clf = SVC()

svc_clf.fit(x_train_scaled, y_train)
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
import numpy as np
cross_score = cross_val_score(svc_clf, x_train_scaled, y_train, cv=5)
print(f'Mean Score {np.mean(cross_score)}')
```

```python
svc_clf.score(x_test_scaled, y_test) 
```

```python

y_pred_svc = svc_clf.predict(x_test_scaled)

print(classification_report(y_test, y_pred_svc))
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train_scaled, y_train)
```

```python
print("Naive Bayes score: ",nb.score(x_test_scaled, y_test))
```

```python
y_pred_gnb = nb.predict(x_test_scaled)

print(classification_report(y_test, y_pred_gnb))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train_scaled,y_train)
```

```python
y_pred_dtc = clf.predict(x_test_scaled)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_dtc))
```

```python
print(classification_report(y_test, y_pred_dtc))
```

### KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train_scaled, y_train)
```

```python
y_pred_knn = knn.predict(x_test_scaled)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_knn))
```

```python
print(classification_report(y_test, y_pred_knn))
```
